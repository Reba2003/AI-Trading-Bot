```python
from openai import OpenAI
import os


client = OpenAI(api_key="OPENAI_API_KEY")

def get_ai_risk_analysis(portfolio_data):
    """
    Sends portfolio state to GPT-4o for a risk heuristic analysis.
    """
    system_prompt = {
        "role": "system", 
        "content": "You are a Quantitative Risk Manager. Analyze the provided portfolio for tail risk and concentration."
    }
    user_message = {
        "role": "user", 
        "content": f"Current Portfolio State: {portfolio_data}. Provide a concise risk assessment."
    }
    
    try:
        response = client.chat.completions.create(
            model="gpt-4o", # or "gpt-4o-mini" for lower cost
            messages=[system_prompt, user_message],
            temperature=0.7
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"Risk Analysis Error: {str(e)}"
```
